# 2. Load barcode metadata and add to obs and vars

```python
meta_folder = r'L:\Shiwei\DNA_MERFISH_analysis\Paired_tag\GSE152020_Paired-Tag_RNA_filtered_matrix'
meta_fname = os.path.join(meta_folder, r'Nuclei_metaData.csv')
meta_df = pd.read_csv(meta_fname,sep='\t',)
#meta_df = meta_df.set_index('Cell_ID', drop=True)
meta_df 
```

```python
print(np.unique(meta_df['Tissue']))
print(np.unique(meta_df['Annotation']))

sel_meta_df = meta_df[meta_df['Tissue']=='FC']
sel_class_list = ['BR_InNeu_CGE', 'BR_InNeu_Pvalb', 'BR_InNeu_Sst', 'BR_NonNeu_Astro_Myoc',
 'BR_NonNeu_Astro_Nnat', 'BR_NonNeu_Endothelial', 'BR_NonNeu_Microglia',
 'BR_NonNeu_OPC', 'BR_NonNeu_Oligo_MFOL', 'BR_NonNeu_Oligo_MOL',
 'FC_ExNeu_CT','FC_ExNeu_L23', 'FC_ExNeu_L4',
 'FC_ExNeu_L5', 'FC_ExNeu_NP', 'FC_ExNeu_PT' ]
sel_meta_df = sel_meta_df[sel_meta_df['Annotation'].isin(sel_class_list)]


sel_meta_df.set_index('Cell_ID', inplace=True)
sel_meta_df
```

```python
print(np.unique(meta_df['Target']))
target_mode = 'H3K27me3'
sel_meta_df_histone = sel_meta_df[sel_meta_df['Target']==target_mode]
sel_meta_df_histone
```

# 3. Convert selected obs to annData

```python
ann_df = df.T 
del df # free memory
#ann_df.head()

# select the desired obs
sel_ann_df = ann_df.loc[sel_meta_df_histone.index]
del ann_df # free memory

sel_ann_df
```

## make annData

```python
adata = ad.AnnData(sel_ann_df)
adata
```

## add obs

```python
adata.obs = sel_meta_df_histone.copy(deep=True)
adata.obs.head()
```

## add vars

```python
loci_names = sel_ann_df.columns
loci_names = [i.replace(':','_').replace('-','_') for i in loci_names]
adata.var_names = loci_names
adata.var.head()
```

## use consistent cell labels

```python
adata.obs['cluster'] =  adata.obs['Annotation']

celltype_rename_dict = {'BR_NonNeu_Endothelial':'Endo', 
                        'BR_NonNeu_Microglia':'Micro', 
                        'BR_NonNeu_Oligo_MFOL':'Oligo',
                        'BR_NonNeu_Oligo_MOL':'Oligo',
                        'BR_NonNeu_Astro_Myoc':'Astro',
                        'BR_NonNeu_Astro_Nnat':'Astro',
                        'BR_NonNeu_OPC':'OPC', 
                        'BR_InNeu_CGE':'CGE',
                        'BR_InNeu_Pvalb':'Pvalb',
                        'BR_InNeu_Sst':'Sst',
                        'FC_ExNeu_CT':'L6 CT',
                        'FC_ExNeu_L23':'L2/3 IT',
                        'FC_ExNeu_L4':'L4/5 IT',
                        'FC_ExNeu_L5':'L5 IT',
                        #'FC_ExNeu_L6':'L6 IT',
                        'FC_ExNeu_NP':'L5/6 NP',
                        'FC_ExNeu_PT':'L5 ET',}
new_class_label_dict = {'cluster':celltype_rename_dict}
adata.obs = adata.obs.replace(new_class_label_dict)
```

```python
adata.obs
```

## convert to sparse matrix

```python
import scipy
adata.X = scipy.sparse.csr_matrix(adata.X)
```
