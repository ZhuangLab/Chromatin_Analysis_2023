# 2. Assemble mean hic matrix for celltype of interest

## process

```python
hic_main_folder = r'\\10.245.74.212\Chromatin_NAS_2\m3C_mouse_brain\isocortex'

hic_tsv_folder = os.path.join(hic_main_folder, r'download')
hic_matrix_folder = os.path.join(hic_main_folder, r'matrix')
```

```python
all_sorted_chr = ['chr' + str(i) for i in range(1,20)]
all_sorted_chr.append('chrX')
all_sorted_chr[:2]
```

```python
import pickle
import tqdm

hic_matrix_ensemble_dict_by_subclass = {}
hic_info_ensemble_dict_by_subclass = {}

for _celltype, _hic_celltype in tqdm.tqdm(celltype_map_dict.items()):
    print (f'Process {_celltype}', end=' ')

    cell_ic_meta_df = ic_meta_df[ic_meta_df['Subclass']==_hic_celltype]
    cell_names = cell_ic_meta_df['cell'].tolist()

    hic_matrix_fname_list = [os.path.join(hic_matrix_folder, f'{cell_name}.3C.contact.matrix.pkl') 
                             for cell_name in cell_names]
    
    hic_matrix_ensemble_dict = {}
    hic_info_ensemble_dict = {}
    
    #for chosen_chrom in ['chr1']:
    for chosen_chrom in all_sorted_chr:
        hic_matrix_chr_list = []
        # append each single cell for chosen chrom
        for hic_matrix_fname in hic_matrix_fname_list:
            if os.path.exists(hic_matrix_fname):
                hic_matrix_dict = pickle.load(open(hic_matrix_fname,'rb'))
                hic_matrix_chr = hic_matrix_dict[chosen_chrom]
                hic_matrix_chr_list.append(hic_matrix_chr)
        # calculate cell type average: use mean here since contacts are sparse 
        # median would be 0 for long-range
        hic_matrix_chr_arr = np.array(hic_matrix_chr_list)
        #print (f'recovered {len(hic_matrix_chr_arr)} cells.')
        hic_matrix_ensemble_dict[chosen_chrom] = np.nanmean(hic_matrix_chr_arr, axis=0)
        hic_info_ensemble_dict[chosen_chrom] = len(hic_matrix_chr_arr)
    
    print (f"recovered {hic_info_ensemble_dict['chr1']} cells.")
    hic_matrix_ensemble_dict_by_subclass[_celltype] = hic_matrix_ensemble_dict
    hic_info_ensemble_dict_by_subclass[_celltype] = {'m3C cell type':celltype_map_dict[_celltype], 
                                                     'cell number':hic_info_ensemble_dict}
```

## save 

```python
# save the dict if needed
import pickle
hic_matrix_dict_savename = os.path.join(meta_folder, 'm3c_isocortex_hic_matrix_by_subclass.pkl')
pickle.dump(hic_matrix_ensemble_dict_by_subclass, open(hic_matrix_dict_savename,'wb'))

hic_info_dict_savename = os.path.join(meta_folder, 'm3c_isocortex_hic_info_by_subclass.pkl')
pickle.dump(hic_info_ensemble_dict_by_subclass, open(hic_info_dict_savename,'wb'))
```
